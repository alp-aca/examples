# Tutorial to obtain 1 Wilson coefficient (WC) exclusion plots
In this tutorial we deal with the obtaintion of the exclusion plot for the ALP EFT, one WC at a time performed in Section 6.2 of our work <a href="#AlpacaPheno">[0]</a>.

```python
import alpaca
from alpaca import ALPcouplings

import numpy as np
import matplotlib.pyplot as plt

from alpaca.statistics import get_chi2
from alpaca.sectors import default_sectors
from alpaca.plotting.mpl import exclusionplot

plt.rcParams.update({'font.size': 12, 'text.usetex': True, 'font.family': 'serif', 'font.serif': 'Computer Modern Roman'})
```

```python
ma_vec=np.logspace(-2, 1, 100)  # axion mass in GeV
 
fa_vec_1=np.logspace(1,7,100)
fa_vec_2=np.logspace(1,7,100)

fa_1TeV=1e3
couplings=np.logspace(-4, 2, 100)

x_ma_1, y_fa_1=np.meshgrid(ma_vec, fa_vec_1)
x_ma_2, y_fa_2=np.meshgrid(ma_vec, fa_vec_2)
```

## Running and matching of the coefficients

### $c_q$

```python
couplings_cq_1= [ALPcouplings({'cqL' :1.0}, scale= 4*np.pi*fa, basis='derivative_above').match_run(10.01, 'VA_below', integrator='leadinglog') for fa in fa_vec_1]
couplings_cq_2= [ALPcouplings({'cqL' :c}, scale= 4*np.pi*fa_1TeV, basis='derivative_above').match_run(10.01, 'VA_below', integrator='leadinglog') for c in couplings]
```

### $c_u$

```python
couplings_cu_1= [ALPcouplings({'cuR' :1.0}, scale= 4*np.pi*fa, basis='derivative_above').match_run(10.01, 'VA_below', integrator='leadinglog') for fa in fa_vec_1] 
couplings_cu_2= [ALPcouplings({'cuR' :c}, scale= 4*np.pi*fa_1TeV, basis='derivative_above').match_run(10.01, 'VA_below', integrator='leadinglog') for c in couplings]
```

### $c_d$

```python
couplings_cd_1= [ALPcouplings({'cdR' :1.0}, scale= 4*np.pi*fa, basis='derivative_above').match_run(10.01, 'VA_below', match_tildecouplings=True) for fa in fa_vec_1]
couplings_cd_2= [ALPcouplings({'cdR' :c}, scale= 4*np.pi*fa_1TeV, basis='derivative_above').match_run(10.01, 'VA_below', match_tildecouplings=True) for c in couplings]
```

### $c_l$

```python
couplings_cl_1= [ALPcouplings({'clL' :1.0}, scale= 4*np.pi*fa, basis='derivative_above').match_run(10.01, 'VA_below', match_tildecouplings=True) for fa in fa_vec_2]
couplings_cl_2= [ALPcouplings({'clL' :c}, scale= 4*np.pi*fa_1TeV, basis='derivative_above').match_run(10.01, 'VA_below', match_tildecouplings=True) for c in couplings]
```

### $c_e$

```python
couplings_ce_1= [ALPcouplings({'ceR' :1.0}, scale= 4*np.pi*fa, basis='derivative_above').match_run(10.01, 'VA_below', match_tildecouplings=True) for fa in fa_vec_2]
couplings_ce_2= [ALPcouplings({'ceR' :c}, scale= 4*np.pi*fa_1TeV, basis='derivative_above').match_run(10.01, 'VA_below', match_tildecouplings=True) for c in couplings]
```

### $c_G$

```python
couplings_cG_1= [ALPcouplings({'cG' :1.0}, scale= 4*np.pi*fa, basis='derivative_above').match_run(10.01, 'VA_below', match_tildecouplings=True) for fa in fa_vec_1] 
couplings_cG_2= [ALPcouplings({'cG' :c}, scale= 4*np.pi*fa_1TeV, basis='derivative_above').match_run(10.01, 'VA_below', match_tildecouplings=True) for c in couplings]
```

### $c_B$

```python
couplings_cB_1= [ALPcouplings({'cB' :1.0}, scale= 4*np.pi*fa, basis='derivative_above').match_run(10.01, 'VA_below', match_tildecouplings=True) for fa in fa_vec_2]
couplings_cB_2= [ALPcouplings({'cB' :c}, scale= 4*np.pi*fa_1TeV, basis='derivative_above').match_run(10.01, 'VA_below', match_tildecouplings=True) for c in couplings]
```

### $c_W$

```python
couplings_cW_1= [ALPcouplings({'cW' :1.0}, scale= 4*np.pi*fa, basis='derivative_above').match_run(10.01, 'VA_below', match_tildecouplings=True) for fa in fa_vec_2]
couplings_cW_2= [ALPcouplings({'cW' :c}, scale= 4*np.pi*fa_1TeV, basis='derivative_above').match_run(10.01, 'VA_below', match_tildecouplings=True) for c in couplings]
```

## $\chi^2$ calculation

### $c_q$

```python
x_ma, couplings_cq_grid_1 = np.meshgrid(ma_vec, couplings_cq_1)
x_ma, couplings_cq_grid_2 = np.meshgrid(ma_vec, couplings_cq_2)
chi2_cq_1=get_chi2(default_sectors['all'], x_ma, couplings_cq_grid_1, y_fa_1, integrator='no_rge', min_probability=0)
chi2_cq_2=get_chi2(default_sectors['all'], x_ma, couplings_cq_grid_2, fa_1TeV, integrator='no_rge', min_probability=0)
```

### $c_u$

```python
x_ma, couplings_cu_grid_1 = np.meshgrid(ma_vec, couplings_cu_1)
x_ma, couplings_cu_grid_2 = np.meshgrid(ma_vec, couplings_cu_2)
chi2_cu_1=get_chi2(default_sectors['all'], x_ma, couplings_cu_grid_1, y_fa_1, integrator='no_rge', min_probability=0)
chi2_cu_2=get_chi2(default_sectors['all'], x_ma, couplings_cu_grid_2, fa_1TeV, integrator='no_rge', min_probability=0)
```

### $c_d$

```python
x_ma, couplings_cd_grid_1 = np.meshgrid(ma_vec, couplings_cd_1)
x_ma, couplings_cd_grid_2 = np.meshgrid(ma_vec, couplings_cd_2)
chi2_cd_1=get_chi2(default_sectors['all'], x_ma, couplings_cd_grid_1, y_fa_1, integrator='no_rge', min_probability=0)
chi2_cd_2=get_chi2(default_sectors['all'], x_ma, couplings_cd_grid_2, fa_1TeV, integrator='no_rge', min_probability=0)
```

### $c_l$

```python
x_ma, couplings_cl_grid_1 = np.meshgrid(ma_vec, couplings_cl_1)
x_ma, couplings_cl_grid_2 = np.meshgrid(ma_vec, couplings_cl_2)
chi2_cl_1=get_chi2(default_sectors['all'], x_ma, couplings_cl_grid_1, y_fa_2, integrator='no_rge', min_probability=0)
chi2_cl_2=get_chi2(default_sectors['all'], x_ma, couplings_cl_grid_2, fa_1TeV, integrator='no_rge', min_probability=0)
```

### $c_e$

```python
x_ma, couplings_ce_grid_1 = np.meshgrid(ma_vec, couplings_ce_1)
x_ma, couplings_ce_grid_2 = np.meshgrid(ma_vec, couplings_ce_2)
chi2_ce_1=get_chi2(default_sectors['all'], x_ma, couplings_ce_grid_1, y_fa_2, integrator='no_rge', min_probability=0)
chi2_ce_2=get_chi2(default_sectors['all'], x_ma, couplings_ce_grid_2, fa_1TeV, integrator='no_rge', min_probability=0)
```

### $c_G$

```python
x_ma, couplings_cG_grid_1 = np.meshgrid(ma_vec, couplings_cG_1)
x_ma, couplings_cG_grid_2 = np.meshgrid(ma_vec, couplings_cG_2)
chi2_cG_1=get_chi2(default_sectors['all'], x_ma, couplings_cG_grid_1, y_fa_1, integrator='no_rge', min_probability=0)
chi2_cG_2=get_chi2(default_sectors['all'], x_ma, couplings_cG_grid_2, fa_1TeV, integrator='no_rge', min_probability=0)
```

### $c_B$

```python
x_ma, couplings_cB_grid_1 = np.meshgrid(ma_vec, couplings_cB_1)
x_ma, couplings_cB_grid_2 = np.meshgrid(ma_vec, couplings_cB_2)
chi2_cB_1=get_chi2(default_sectors['all'], x_ma, couplings_cB_grid_1, y_fa_2, integrator='no_rge', min_probability=0)
chi2_cB_2=get_chi2(default_sectors['all'], x_ma, couplings_cB_grid_2, fa_1TeV, integrator='no_rge', min_probability=0)
```

### $c_W$

```python
x_ma, couplings_cW_grid_1 = np.meshgrid(ma_vec, couplings_cW_1)
x_ma, couplings_cW_grid_2 = np.meshgrid(ma_vec, couplings_cW_2)
chi2_cW_1=get_chi2(default_sectors['all'], x_ma, couplings_cW_grid_1, y_fa_2, integrator='no_rge', min_probability=0)
chi2_cW_2=get_chi2(default_sectors['all'], x_ma, couplings_cW_grid_2, fa_1TeV, integrator='no_rge', min_probability=0)
```

## Plot

Plotting of the $\chi^2$ significance

```python
from matplotlib.lines import Line2D

fig, ax =plt.subplots(1, 2, sharey=True, figsize=(12, 6))
x_ma_c, coups_grid= np.meshgrid(ma_vec, couplings)

# Filled contours of the 3 sigma exclusion levels
cs_cq_1=ax[0].contourf(x_ma, 1/y_fa_1, chi2_cq_1[0].significance(), levels=[3,100], colors='red', alpha=0.1)
cs_cu_1=ax[0].contourf(x_ma, 1/y_fa_1, chi2_cu_1[0].significance(), levels=[3,100], colors='blue', alpha=0.1)
cs_cG_1=ax[0].contourf(x_ma, 1/y_fa_1, chi2_cG_1[0].significance(), levels=[3,100], colors='green', alpha=0.1)
cs_cd_1=ax[0].contourf(x_ma, 1/y_fa_1, chi2_cd_1[0].significance(), levels=[3,100], colors='orange', alpha=0.1)
cs_cq_1=ax[0].contour(x_ma, 1/y_fa_1, chi2_cq_1[0].significance(), levels=[3], colors='red')
cs_cu_1=ax[0].contour(x_ma, 1/y_fa_1, chi2_cu_1[0].significance(), levels=[3], colors='blue')
cs_cG_1=ax[0].contour(x_ma, 1/y_fa_1, chi2_cG_1[0].significance(), levels=[3], colors='green')
cs_cd_1=ax[0].contour(x_ma, 1/y_fa_1, chi2_cd_1[0].significance(), levels=[3], colors='orange')
cs_cq_2=ax[0].contour(x_ma, coups_grid/fa_1TeV, chi2_cq_2[0].significance(), levels=[3], colors='red', linestyles='dashed')
cs_cu_2=ax[0].contour(x_ma, coups_grid/fa_1TeV, chi2_cu_2[0].significance(), levels=[3], colors='blue', linestyles='dashed')
cs_cG_2=ax[0].contour(x_ma, coups_grid/fa_1TeV, chi2_cG_2[0].significance(), levels=[3], colors='green', linestyles='dashed')
cs_cd_2=ax[0].contour(x_ma, coups_grid/fa_1TeV, chi2_cd_2[0].significance(), levels=[3], colors='orange', linestyles='dashed')

cs_cl_1=ax[1].contourf(x_ma, 1/y_fa_2, chi2_cl_1[0].significance(), levels=[3,10000], colors='purple', alpha=0.1)
cs_ce_1=ax[1].contourf(x_ma, 1/y_fa_2, chi2_ce_1[0].significance(), levels=[3,10000], colors='brown', alpha=0.1)
cs_cW_1=ax[1].contourf(x_ma, 1/y_fa_2, chi2_cW_1[0].significance(), levels=[3,10000], colors='magenta', alpha=0.1)
cs_cl_1=ax[1].contour(x_ma, 1/y_fa_2, chi2_cl_1[0].significance(), levels=[3], colors='purple')
cs_ce_1=ax[1].contour(x_ma, 1/y_fa_2, chi2_ce_1[0].significance(), levels=[3], colors='brown')
cs_cW_1=ax[1].contour(x_ma, 1/y_fa_2, chi2_cW_1[0].significance(), levels=[3], colors='magenta')
cs_cl_2=ax[1].contour(x_ma, coups_grid/fa_1TeV, chi2_cl_2[0].significance(), levels=[3], colors='purple', linestyles='dashed')
cs_ce_2=ax[1].contour(x_ma, coups_grid/fa_1TeV, chi2_ce_2[0].significance(), levels=[3], colors='brown', linestyles='dashed')
cs_cW_2=ax[1].contour(x_ma, coups_grid/fa_1TeV, chi2_cW_2[0].significance(), levels=[3], colors='magenta', linestyles='dashed')

ax[0].set_xscale('log')
ax[0].set_yscale('log')

ax[1].set_xscale('log')
ax[1].set_yscale('log')

ax[0].set_ylim(1e-7, 1e-1)


ax[0].set_xlabel(r'$m_a$ [GeV]')

ax[1].set_xlabel(r'$m_a$ [GeV]')
ax[0].set_ylabel(r'$c_i/f_a$ [GeV]')

handles= [ Line2D([0], [0], color='red', lw=2),
           Line2D([0], [0], color='blue', lw=2),
           Line2D([0], [0], color='orange', lw=2),
           Line2D([0], [0], color='green', lw=2)]
labels = [r'$c_{q_L}$', r'$c_{u_R}$', r'$c_{d_R}$', r'$c_{G}$']
ax[0].legend(handles, labels, loc='upper left', fontsize=12)
handles= [ Line2D([0], [0], color='purple', lw=2),
           Line2D([0], [0], color='brown', lw=2),
           # Line2D([0], [0], color='cyan', lw=2),
           Line2D([0], [0], color='magenta', lw=2)]
labels = [r'$c_{l_L}$', r'$c_{e_R}$',
          #r'$c_{cB}$', 
          r'$c_{W}$']
ax[1].legend(handles, labels, loc='upper left', fontsize=12)

fig.tight_layout()
#fig.savefig('ALP_couplings_contours.pdf', bbox_inches='tight')
```
